9. Text-Based Decision Game
Create a text-based decision-making game.
The game should present players with scenarios where choices affect the storyline and
score. Multiple story paths and endings must exist.
The system should track decisions, compute scores, and allow replayability. String
processing and conditional branching are core requirements.


```python
def play_game():
    score = 0
    decisions = []

    print("\n" + "=" * 50)
    print("     WELCOME TO THE HAUNTED FOREST")
    print("=" * 50)

    name = input("Enter your name: ")
    print(f"\nHello {name}! Your adventure begins now.")


    print("\nYou arrive at the entrance of a haunted forest.")
    print("A. Follow the mysterious path")
    print("B. Enter the dark cave")


    choice1 = input("Enter your choice (A/B): ").upper()

    if choice1 == "A":
        score += 20
        decisions.append("Followed the mysterious path")


        print("\nYou reach a river.")
        print("A. Cross the river")
        print("B. Walk along the river bank")

        choice2 = input("Enter your choice (A/B): ").upper()

        if choice2 == "A":
            score += 20
            decisions.append("Crossed the river")


            print("\nYou find a trapped traveler.")
            print("A. Help the traveler")
            print("B. Ignore and continue")

            choice3 = input("Enter your choice (A/B): ").upper()

            if choice3 == "A":
                score += 30
                decisions.append("Helped the traveler")
                ending = "HERO ENDING 🏆"

            else:
                score -= 8
                decisions.append("Ignored the traveler")
                ending = "LONELY ENDING 😔"

        elif choice2 == "B":
            score += 10
            decisions.append("Walked along the river")

            print("\nYou get lost in the forest.")
            ending = "LOST ENDING 🌲"

        else:
            print("Invalid choice!")
            return

    elif choice1 == "B":
        score += 8
        decisions.append("Entered the dark cave")


        print("\nInside the cave, you discover a treasure chest.")
        print("A. Open the chest")
        print("B. Leave it alone")

        choice2 = input("Enter your choice (A/B): ").upper()

        if choice2 == "A":
            score += 20
            decisions.append("Opened the treasure chest")

            print("\nThe chest contains gold and a magical map.")
            print("A. Take the gold")
            print("B. Take the magical map")

            choice3 = input("Enter your choice (A/B): ").upper()

            if choice3 == "A":
                score += 20
                decisions.append("Took the gold")
                ending = "RICH ENDING 💰"


            else:
                score += 30
                decisions.append("Took the magical map")
                ending = "EXPLORER ENDING 🗺️"

        elif choice2 == "B":
            score -= 5
            decisions.append("Ignored the treasure chest")
            ending = "REGRET ENDING 😢"

        else:
            print("Invalid choice!")
            return

    else:
        print("Invalid choice!")
        return


    print("\n" + "=" * 50)
    print("              GAME OVER")
    print("=" * 50)

    print(f"\nPlayer: {name}")
    print(f"Ending Achieved: {ending}")
    print(f"Final Score: {score}")


    if score >= 50:
        rank = "Master Adventurer"
    elif score >= 30:
        rank = "Expert Explorer"
    elif score >= 20:
        rank = "Adventurer"
    else:
        rank = "Beginner"

    print(f"Rank: {rank}")

    print("\nYour Decisions:")
    for i, decision in enumerate(decisions, start=1):
        print(f"{i}. {decision}")

    print("\nThank you for playing!")



while True:
    play_game()

    replay = input("\nDo you want to play again? (yes/no): ").lower()

    if replay != "yes":
        print("\nGoodbye! See you next time.")
        break
```
