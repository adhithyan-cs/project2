```python


inventory = {}

TAX_RATE = 0.05

def add_product():
    name = input("Enter product name: ")

    try:
        stock = int(input("Enter stock quantity: "))
        price = float(input("Enter price per unit: "))

        if stock < 0 or price < 0:
            print("Stock and price cannot be negative.")
            return

        inventory[name] = {"stock": stock, "price": price}
        print("Product added successfully!")

    except ValueError:
        print("Invalid input!")


def update_product():
    name = input("Enter product name to update: ")

    if name not in inventory:
        print("Product not found.")
        return

    try:
        stock = int(input("Enter new stock quantity: "))
        price = float(input("Enter new price: "))

        if stock < 0 or price < 0:
            print("Negative values are not allowed.")
            return

        inventory[name]["stock"] = stock
        inventory[name]["price"] = price

        print("Product updated successfully!")

    except ValueError:
        print("Invalid input!")


def display_inventory():
    if not inventory:
        print("Inventory is empty.")
        return

    print("\n------ INVENTORY ------")
    print(f"{'Product':15}{'Stock':10}{'Price'}")

    for product, details in inventory.items():
        print(
            f"{product:15}{details['stock']:10}{details['price']:.2f}"
        )

    print()


def calculate_discount(amount):
    """
    Discount Slabs:
    1000 - 1999 : 5%
    2000 - 4999 : 10%
    >= 5000     : 15%
    """

    if amount >= 5000:
        return amount * 0.15
    elif amount >= 2000:
        return amount * 0.10
    elif amount >= 1000:
        return amount * 0.05
    return 0


def purchase_products():
    if not inventory:
        print("No products available.")
        return

    cart = []

    while True:
        product = input("\nEnter product name: ")

        if product not in inventory:
            print("Product not found.")
            continue

        try:
            qty = int(input("Enter quantity: "))

            if qty <= 0:
                print("Quantity must be greater than zero.")
                continue

            if qty > inventory[product]["stock"]:
                print("Insufficient stock available.")
                continue

            cart.append((product, qty))

            choice = input("Add more items? (y/n): ").lower()
            if choice != 'y':
                break

        except ValueError:
            print("Invalid quantity.")

    generate_bill(cart)


def generate_bill(cart):
    subtotal = 0

    print("\n" + "=" * 50)
    print("            CAMPUS STORE BILL")
    print("=" * 50)
    print(f"{'Item':15}{'Qty':8}{'Price':10}{'Amount'}")

    for product, qty in cart:
        price = inventory[product]["price"]
        amount = qty * price
        subtotal += amount

        print(
            f"{product:15}{qty:<8}{price:<10.2f}{amount:.2f}"
        )
        inventory[product]["stock"] -= qty

    tax = subtotal * TAX_RATE
    discount = calculate_discount(subtotal)

    final_amount = subtotal + tax - discount

    print("-" * 50)
    print(f"Subtotal : {subtotal:.2f}")
    print(f"Tax (5%) : {tax:.2f}")
    print(f"Discount : {discount:.2f}")
    print(f"Payable Amount : {final_amount:.2f}")
    print("=" * 50)

    low_stock_alert()


def low_stock_alert():
    print("\nLOW STOCK ALERTS")

    alert_found = False

    for product, details in inventory.items():
        if details["stock"] <= 5:
            print(f"{product} stock is low ({details['stock']} left)")
            alert_found = True

    if not alert_found:
        print("No low-stock items.")

def main():
    while True:
        print("\n===== CAMPUS STORE SYSTEM =====")
        print("1. Add Product")
        print("2. Update Product")
        print("3. View Inventory")
        print("4. Purchase Products")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_product()

        elif choice == "2":
            update_product()

        elif choice == "3":
            display_inventory()

        elif choice == "4":
            purchase_products()

        elif choice == "5":
            print("Thank you!")
            break

        else:
            print("Invalid choice.")


main()
```
